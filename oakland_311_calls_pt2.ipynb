Import the clean version of oakland 311 calls data and set up dtypes.

```python
oakland_311 = pd.read_csv('oakland_311_clean.csv',
    dtype = {
        'REQUESTID': object,
    },
    parse_dates = ['DATETIMEINIT', 'DATETIMECLOSED', 'Close_Time']                       
)
```

```python
oakland_311_2020
```

### Count how many indidents per year

Check the minimum date and maximum date of oakland_311

```python
oakland_311['DATETIMEINIT'].min()
```

```python
oakland_311['DATETIMEINIT'].max()
```

Because the data for 2009 and 2021 is not complete, will need to create a new dataframe with values only from 2010 to 2020.

```python
oakland_311_complete = oakland_311[

    (oakland_311['DATETIMEINIT'] >= '2010-01-01') &
    (oakland_311['DATETIMEINIT'] < '2021-01-01')

].reset_index(drop = True)
```

#### A new dataframe for annual cases

Group by with datetimes

```python
oakland_311_complete.groupby([pd.Grouper(key = 'DATETIMEINIT', axis = 0, freq = 'A')])
```

Create a new dataframe called annual_cases

```python
annual_cases = oakland_311_complete.groupby([pd.Grouper(key = 'DATETIMEINIT', axis = 0, freq = 'A')]).count()
```

```python
annual_cases
```

#### Subset just the REQUESTID column from it

```python
annual_cases = annual_cases[['REQUESTID']].copy()
```

```python
annual_cases
```

Reset the index.

```python
annual_cases = annual_cases.reset_index()
```

```python
annual_cases
```

Rename columns.

```python
annual_cases.rename(columns = {'REQUESTID': 'CASE_COUNT'}, inplace = True)
annual_cases
```

#### Reframe the table

Create a new column called YEAR

```python
annual_cases['YEAR'] = annual_cases['DATETIMEINIT'].dt.year
annual_cases
```

Subset the dataframe with just the YEAR and CASE_COUNT columns

```python
annual_cases = annual_cases[['YEAR', 'CASE_COUNT']].copy()
annual_cases
```

#### Visualize the table

```python
alt.Chart(annual_cases).mark_bar().encode(
   
    x = 'YEAR:O',
    y = 'CASE_COUNT'
    
).properties(
    
    title = 'Oakland 311 calls: Number of cases by year'
    
)
```

### Median Close_Time by year

Get the median length of time it takes to close a case by year.

Create a new column to onvert Close_Time to seconds.

```python
oakland_311_complete['CLOSE_TIME_SECONDS'] = oakland_311_complete['Close_Time'].dt.total_seconds()
```

Check the result.

```python
oakland_311_complete.head()
```

Try to get the values of median length of Close_Time

```python
median_close_time = oakland_311_complete.groupby(
    
    [pd.Grouper(key = 'DATETIMEINIT', axis = 0, freq = 'A'),]

).median()
```

```python
median_close_time
```

Subset just the CLOSE_TIME_SECONDS from it.

```python
median_close_time = median_close_time[['CLOSE_TIME_SECONDS']].reset_index()
median_close_time
```

Create a new column for years.

```python
median_close_time['YEAR'] = median_close_time['DATETIMEINIT'].dt.year
```

```python
median_close_time
```

Convert seconds to days.

```python
median_close_time['CLOSE_TIME_DAYS'] = median_close_time['CLOSE_TIME_SECONDS'] / 60 / 60 / 24
```

```python
median_close_time
```

Subset the dataframe with only YEAR and CLOSE_TIME_DAYS columns

```python
median_close_time = median_close_time[['YEAR', 'CLOSE_TIME_DAYS']].copy()
median_close_time
```

Visualize the table.

```python
alt.Chart(median_close_time[['YEAR', 'CLOSE_TIME_DAYS']]).mark_bar().encode(
    
    x = 'YEAR:O',
    y = 'CLOSE_TIME_DAYS',
    
).properties(
    
    title = 'Oakland 311 calls: Median resolution time'
    
)
```

### Merge two dataframes

Merge annual_cases and median_close_time.

```python
annual_cases
```

```python
median_close_time
```

Combine them into a new table called annual_summary, using outer.

```python
annual_summary = pd.merge(
    
    annual_cases,
    median_close_time,
    on = 'YEAR',
    how = 'outer',
    validate = '1:1'
    
)
```

### Number of illegal dumping cases per year

When analyzing categories of incidents in 2020, illegal dumping topped the list. So I think it's worth looking into it.

Creating a dataframe with illegal dumping cases in 2010 to 2020 only. 

```python
oakland_311_illdump = oakland_311_complete[

    (oakland_311_complete['REQCATEGORY'] == 'ILLDUMP')

].reset_index(drop = True)
```

Take a look at the new dataframe.

```python
oakland_311_illdump
```

Create a YEAR column for the dataframe.

```python
oakland_311_illdump['YEAR'] = oakland_311_illdump['DATETIMEINIT'].dt.year
oakland_311_illdump
```

Group by year and create a new dataframe.

```python
illdump_counts = oakland_311_illdump.groupby(['YEAR']).count()
illdump_counts
```

Subset just the one column, REQUESTID, from illdump_counts

```python
illdump_counts = illdump_counts[['REQUESTID']].copy()
illdump_counts
```

Reset the index.

```python
illdump_counts = illdump_counts.reset_index()
illdump_counts
```

Rename the column.

```python
illdump_counts.rename(columns = {'REQUESTID': 'COUNT'}, inplace = True)
illdump_counts
```

Visualize the table.

```python
alt.Chart(illdump_counts).mark_bar().encode(
    
    x = 'YEAR',
    y = 'COUNT'
    
).properties(
    
    title = 'Oakland 311 illegal dumping cases per year'
    
)
```
